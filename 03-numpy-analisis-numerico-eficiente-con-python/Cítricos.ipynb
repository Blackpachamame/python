# **Numpy: Análisis numérico eficiente con Python**

## Importando las bibliotecas

```python
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/Blackpachamame/python/refs/heads/main/03-numpy-analisis-numerico-eficiente-con-python/citricos.csv'
datos = np.loadtxt(url, delimiter=',', usecols=np.arange(1, 6, 1), skiprows=1)
```

## Exploración de Datos

Debes seleccionar parte de los datos. Las columnas que evaluaremos son el diámetro y el peso. Crea arrays específicos para almacenar el diámetro y el peso de la naranja y la toronja. El diámetro está en la columna cero y el peso en la columna uno. Los datos de las naranjas van hasta la fila 4999 y los datos de las toronjas comienzan en la fila 5000 del archivo.

Después de seleccionar los datos, importa la biblioteca matplotlib y crea un gráfico para el peso en función del diámetro tanto para las naranjas como para las toronjas.

```python
diametro_naranja = datos[:5000, 0]
diametro_toronja = datos[5000:, 0]
peso_naranja = datos[:5000, 1]
peso_toronja = datos[5000:, 1]
```

```python
import matplotlib.pyplot as plt

plt.plot(diametro_naranja, peso_naranja)
plt.title("Peso en función del Diámetro para Naranjas")
plt.xlabel("Diámetro")
plt.ylabel("Peso")
plt.show()
```

```python
plt.plot(diametro_toronja, peso_toronja)
plt.title("Peso en función del Diámetro para Toronjas")
plt.xlabel("Diámetro")
plt.ylabel("Peso")
plt.show()
```

## Operaciones entre Arrays

Ahora debes calcular el coeficiente angular y lineal para la recta de las naranjas y para la recta de las toronjas. Utiliza la fórmula de mínimos cuadrados para encontrar cada uno.

```python
n = np.size(diametro_naranja)
x = diametro_naranja
y = peso_naranja

a = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2)-np.sum(x)**2)
b = np.mean(y) - a*np.mean(x)
```

```python
y = a * x + b
plt.plot(diametro_naranja, peso_naranja)
plt.plot(x, y)
```

```python
n = np.size(diametro_toronja)
x = diametro_toronja
y = peso_toronja

a = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2)-np.sum(x)**2)
b = np.mean(y) - a*np.mean(x)
```

```python
y = a * x + b
plt.plot(diametro_toronja, peso_toronja)
plt.plot(x, y)
```

## Números aleatorios

Debes calcular el coeficiente angular utilizando la generación de números aleatorios. Supongamos que ya conoces el valor de b y que este es igual a 17.

```python
norma = np.array([])
np.random.seed(84)
coeficientes_angulares = np.random.uniform(low=0.0, high=30.0, size=100) # pendiente toronja

for i in range(100):
  norma = np.append(norma, np.linalg.norm(y - (coeficientes_angulares[i]*x + b)))
```
